### 🧠 Why bioinformatics uses the terminal

Bioinformatics tools were built long before graphical interfaces (like Windows) became common. They are designed to be:

- **Fast** — processing millions of reads is not practical by clicking buttons
- **Precise** — commands are exact and reproducible
- **Scriptable** — you can chain commands together to form a pipeline
- **Server-friendly** — most computing clusters only have a terminal, no graphical interface

Because of this, all the tools you will use — `fastqc`, `bwa`, `samtools`, `bcftools` — are run from the terminal.

Linux is the operating system most commonly used on bioinformatics servers. The commands below are standard Linux commands you will see used throughout this course.

## 🔗 The pipe operator `|`

You will see `|` used constantly in the following notebooks. It is one of the most powerful concepts in Linux.

The **pipe** sends the output of one command directly as input to the next:

```
command1 | command2
```

For example, you have already seen this in the course:

```bash
!zcat sample_R1.fastq.gz | head -20
```

- `zcat` decompresses the file and outputs all its lines
- `|` sends those lines to `head`
- `head -20` keeps only the first 20 lines

Try it here — `ls` produces a list of files, and `head -3` keeps only the first 3:

````python
!ls | head -3
````

---

There are dozens of essential Linux commands, hundreds of commonly used ones, and thousands available depending on the system. Nobody uses them all — you learn the ones you need as you go.

The key insight is this: commands can be **combined with pipes** so that, in a single line, you can chain multiple actions together. And they can be written in a text file — called a **script** — where the computer reads and executes each command in order. That is the basis of every bioinformatics pipeline.
