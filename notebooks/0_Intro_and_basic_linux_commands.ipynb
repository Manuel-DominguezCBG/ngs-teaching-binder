## 0. **How does this notebook work?**

This notebook allows us to write plain text, but more importantly, it lets you simulate a Linux terminal.

A terminal is simply a window where you type commands and the computer responds with text. The computer reads the command, executes it, and prints the result. In everyday computer use, you interact with a graphical interface: you click to create folders, open files, save them, move them, delete them, and so on.

In a terminal, you can do exactly the same things — and much more — but using short text commands instead of buttons. It may look unfamiliar at first, but it is extremely powerful and is the standard way bioinformatics tools are used.

🧠 Why bioinformatics uses terminals

Bioinformatics tools were created long before graphical interfaces (e.g., Windows XP or 11) became common. They are designed to be:

- fast  
- precise  
- scriptable (you can automate them)  
- easy to run on servers and clusters  

Because of this, most genomics tools — FastQC, minimap2, samtools, bcftools — are built to run in a terminal.

Windows is actually an operating system. A program that allow you to communicate with the computer. There is another one you may be familiar that is Linux. Linux is more commonly used on servers than Windows because it offers a combination of stability, flexibility, cost efficiency, and performance that fits server workloads exceptionally well.

The `ls` command you execute before is one of many tools provide by Linux to allow you to communicate with the computer.
