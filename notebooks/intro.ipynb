### Why this project exists
When teaching bioinformatics to people without a computational background, I often notice the same reaction: everything feels abstract, mysterious, almost like watching a magician perform a spell. You see the final result, but the steps in between feel hidden inside a computational cauldron.

Bioinformatics is **not** magic. It’s a set of transparent, logical steps that anyone can understand with the right guidance.

To understand what a bioinformatics algorithm does, you don’t need advanced mathematical knowledge (I didn’t even study Maths at A‑levels), and you don’t need to know how a computer works internally, which can be very complex. Bioinformatics tools used in genomics follow simple logical steps. Their complexity comes from the many exceptions and conditions added to handle unusual cases correctly and avoid errors in the analysis. But the core steps are almost always straightforward, and they can be understood. That is exactly what I aim to show you here.

## 0. **How does this notebook work?**

This notebook allows us to write plain text, but more importantly, it lets you simulate a **terminal**. In this training you will be using a **Linux terminal**.

A terminal is simply a window where you type commands and the computer answers back with text. The computer reads the command, executes it, and prints the result. When you normally use a computer, you interact with a **graphical interface**: you click to create a new folder, open a text file, save it, move it, delete it, and so on.

In a terminal you can do exactly the same things — and much more — but using short text commands instead of buttons. It may look unfamiliar at first, but it is extremely powerful and is the standard way bioinformatics tools are used.

Let’s try it together.

`ls`

🧠 Why bioinformatics uses terminals

Bioinformatics tools were created long before graphical interfaces became common. They are designed to be:

- fast  
- precise  
- scriptable (you can automate them)  
- easy to run on servers and clusters  

Because of this, most genomics tools—FastQC, minimap2, samtools, bcftools—are built to run in a terminal.
