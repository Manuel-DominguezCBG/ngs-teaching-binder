## 🧪 Where does NGS data come from?

Before we analyse sequencing data, it is important to understand where it comes from.

Modern sequencing platforms, such as those developed by Illumina, do not “read” DNA in the same way we read text. Instead, they infer the DNA sequence by capturing images of a biochemical process.

---

### 🧬 Step 1 — DNA is fragmented

We start with DNA extracted from a sample (for example, blood or tumour tissue).  

This DNA is too long to sequence directly, so it is broken into many small fragments. These fragments are typically a few hundred base pairs long.

Adapters (short synthetic DNA sequences) are then attached to both ends of each fragment. These adapters allow the fragments to bind to the sequencing flow cell and be amplified.

---

### 🔬 Step 2 — Fragments are attached to a flow cell

The prepared DNA fragments are loaded onto a flow cell — a glass surface coated with oligonucleotides complementary to the adapters.

Each fragment binds to the surface and is amplified locally through a process called **bridge amplification**, creating clusters of identical DNA copies.

Each cluster originates from a single DNA fragment.

---

### ✨ Step 3 — Sequencing by synthesis

Illumina sequencing uses a method called **sequencing by synthesis**.

At each cycle:
- A single base (A, C, G, or T) is incorporated into each DNA cluster
- Each base carries a fluorescent label
- The machine takes an image of the flow cell

The colour detected at each cluster tells the machine which base was added.

This process is repeated cycle by cycle, gradually building up the sequence.

---

### 📸 Step 4 — From images to sequences

The sequencer captures a series of images — one per cycle.

Software then:
- Detects each cluster
- Tracks its signal across cycles
- Converts the sequence of colours into a DNA sequence

This process is called **base calling**.

---

### 📄 Step 5 — FASTQ files are generated

The final output is typically stored in **FASTQ files**.

Each read contains:
- A DNA sequence (the bases that were called)
- A quality score for each base (confidence of the call)

In paired-end sequencing, you will see two files:
- R1 → forward reads  
- R2 → reverse reads  

These are the files we will use throughout this notebook.

---

### 🧠 Key idea

NGS machines do not directly "read" DNA letters.

They:
1. Detect light signals  
2. Convert images into base calls  
3. Output sequences with quality scores  

Understanding this helps explain why:
- Errors occur  
- Quality scores matter  
- Downstream processing (alignment, filtering) is necessary
